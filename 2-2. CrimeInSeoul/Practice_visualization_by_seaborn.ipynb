```python
import numpy as np      # 수치해석/ 통계 관련 라이브러리
import matplotlib.pyplot as plt
# jupyter notebook 명령어
# %matplotlib inline    

# Matplotlib기반의 다양한 색상 테마와 통계용 챠트를 제공하는 시각화 라이브러리(패키지)
import seaborn as sns

# Example
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("whitegrid")
%matplotlib inline    
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.Figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips) # x축 : 요일, y축 : 전체 금액
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
```

```python
sns.set_style('darkgrid')
sns.lmplot(x="total_bill", y="tip", data=tips, size=7)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip",hue="smoker", data=tips, palette="Set1", size=7)
plt.show()
```
